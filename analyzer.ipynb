```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import uncertainties as unc
%matplotlib inline
plt.rcParams["figure.figsize"] = (15, 10)
```

```python
data = pd.read_csv('players_2020.csv')
```

```python
qbs_2020 = data.query('Position == "QB"')
```

```python
rbs_2020 = data.query('Position == "RB"')
```

```python
wrs_2020 = data.query('Position == "WR"')
```

```python
tes_2020 = data.query('Position == "TE"')
```

```python
plt.plot(qbs_2020['Overall Rank'], qbs_2020['Average Points'], 'o')
plt.xlabel('Overall Rank')
plt.ylabel('Average Points')
plt.title('QBs')
plt.show()
```

```python
plt.plot(rbs_2020['Overall Rank'], rbs_2020['Average Points'], 'o')
plt.xlabel('Overall Rank')
plt.ylabel('Average Points')
plt.title('RBs')
plt.show()
```

```python
plt.plot(wrs_2020['Overall Rank'], wrs_2020['Average Points'], 'o')
plt.xlabel('Overall Rank')
plt.ylabel('Average Points')
plt.title('WRs')
plt.show()
```

```python
plt.plot(tes_2020['Overall Rank'], tes_2020['Average Points'], 'o')
plt.xlabel('Overall Rank')
plt.ylabel('Average Points')
plt.title('TEs')
plt.show()
```

```python
amplitudes = []
decays = []
```

```python
from lmfit.models import ExponentialModel
line = ExponentialModel()
fit_x_values = np.linspace(0, max(qbs_2020['Overall Rank']))
param_guess = line.guess(qbs_2020['Average Points'], x=qbs_2020['Overall Rank'])
line_fit_qbs = line.fit(qbs_2020['Average Points'], param_guess, x=qbs_2020['Overall Rank'])
print('Amplitude = ' + str(line_fit_qbs.params['amplitude'].value) + ' +/- ' + str(line_fit_qbs.params['amplitude'].stderr))
print('Decay = ' + str(line_fit_qbs.params['decay'].value) + ' +/- ' + str(line_fit_qbs.params['decay'].stderr))
amplitudes.append(unc.ufloat(line_fit_qbs.params['amplitude'].value, line_fit_qbs.params['amplitude'].stderr))
decays.append(unc.ufloat(line_fit_qbs.params['decay'].value, line_fit_qbs.params['decay'].stderr))
fitline_qbs = line_fit_qbs.eval(line_fit_qbs.params, x=fit_x_values)
plt.plot(fit_x_values, fitline_qbs)
plt.plot(qbs_2020['Overall Rank'], qbs_2020['Average Points'], 'o')
plt.title('Exponential Model Fit of QBs Average Points versus Overall Rank')
plt.xlabel('Overall Rank')
plt.ylabel('Average Points')
plt.show()
```
